```python
def answer_three():
    
    top15 = answer_one()
    date_range = np.arange(2006,2016).astype(str)
    top15['avgGDP'] = top15[date_range].mean(axis=1).dropna()
    top15.reset_index(drop=True)
    top15 = top15.set_index('Country')
    top15.loc['Australia','avgGDP']
    return top15['avgGDP'].sort_values(ascending=False)

answer_three()
```

```python
def answer_four():
    
    top15 = answer_one()
    date_range = np.arange(2006,2016).astype(str)
    top15['avgGDP'] = top15[date_range].mean(axis=1).dropna()
    top15 = top15.sort_values(['avgGDP'],ascending=False)
    top15 = top15.reset_index(drop=True)
    top15['change'] = top15['2015'] - top15['2006']
    return top15.loc[5,'change']

answer_four()
```

```python
def answer_six():
    
    top15 = answer_one()
    top15 = top15.reset_index(drop=True)
    position = top15['% Renewables'].idxmax()
    Country = top15.loc[position,'Country']
    Renewables = top15.loc[position, '% Renewables']
    return (Country , Renewables)

answer_six()
```

```python
def answer_seven():
    
    top15 = answer_one()
    top15
    top15['Citations ratio'] = top15['Self-citations']/top15['Citations']
    position = top15['Citations ratio'].idxmax()
    citations_ratio = top15.loc[position, 'Citations ratio']
    country = top15.loc[position, 'Country']
    return (country, citations_ratio)

answer_seven()
```

```python
def answer_eight():
    
    top15 = answer_one()
    top15['Population'] = top15['Energy Supply']/top15['Energy Supply per Capita']
    top15 = top15.sort_values(['Population'],ascending=False)
    top15 = top15.reset_index(drop=True)
    return top15.loc[2, 'Country']

answer_eight()
```

```python
def answer_nine():
    
    top15 = answer_one()
    top15['Population'] = top15['Energy Supply']/top15['Energy Supply per Capita']
    top15['Citable docs per Capita'] = top15['Citable documents']/top15['Population']
    cols_to_use = (['Citable docs per Capita', 'Energy Supply per Capita'])
    top15 = top15[cols_to_use]
    correlation = top15.corr(method='pearson')
    return correlation.iloc[1]['Citable docs per Capita']

answer_nine()
```

```python
def answer_ten():
    
    top15 = answer_one()
    top15 = top15.reset_index(drop=True)
    top15 = top15.set_index('Country')
    median = top15['% Renewables'].median()
    top15['Above Median'] = (top15['% Renewables']>=median)
    return top15['Above Median']

answer_ten()
```
