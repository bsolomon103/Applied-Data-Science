```python
def get_list_of_university_towns():
    import re 
    import pandas as pd
    import numpy as np 

    file = open ('/Users/emilyvincett/Downloads/university_towns.txt', 'r')
    read = file.readlines() # .readlines breaks the data at the newline


    # To remove the empty spaces
    lines = []
    for line in read:
        if not re.match (r'^\s*$', line):
            lines.append(line)
    lines_copy = lines.copy()

    # To remove the newline '\n' from data
    for index, values in enumerate(lines_copy): 
        lines_copy[index] = values.strip()

    # Create dataframe 
    utowns = pd.DataFrame(columns=('State', 'Region Name'))
    a = 0 
    state = " "
    region = " "

    for line in lines_copy: 
        if '[edit]' in line: 
            state = line.replace ('[edit]', '')
        else: 
            region = re.sub (r' \(.*', " ", line )
            utowns.loc[a] = [state , region]
            a += 1 
    return utowns

get_list_of_university_towns()
   
```

```python
def get_recession_start():
    import pandas as pd 
    import numpy as np 

    gdp = pd.read_excel('/Users/emilyvincett/Downloads/gdplev.xls', skiprows=219, headers=None)
    gdp.drop(['Unnamed: 0', 'Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 7'] ,axis=1,inplace=True)
    gdp.rename ({'1999q4':'Quarter', 9926.1:'GDP', 12323.3:'GDP-2'},axis=1, inplace=True)

    #gdp.head()

    for i in range (4, (len(gdp))): 
        if (gdp.loc[i-4,'GDP']> gdp.loc[i-3,'GDP'])\
            & (gdp.loc[i-3,'GDP']> gdp.loc[i-2,'GDP'])\
            & (gdp.loc[i-2,'GDP']< gdp.loc[i-1,'GDP'])\
            & (gdp.loc[i-1,'GDP']< gdp.loc[i,'GDP']):
            base_year = i-4

    recession_starts = gdp.loc[base_year,'Quarter']

    j = base_year 
    while True:
        if (gdp.loc[j-1,'GDP'] > gdp.loc[j,'GDP']):
            j -= 1
        else:
            recession_starts = gdp.loc[j+1,'Quarter']
            break
    return recession_starts
        
get_recession_start()
```

```python
def convert_housing_data_to_quarters():
    import pandas as pd 
    import numpy as np 
    states = ({'OH': 'Ohio', 'KY': 'Kentucky', 'AS': 'American Samoa', 'NV': 'Nevada', 
          'WY': 'Wyoming', 'NA': 'National', 'AL': 'Alabama', 'MD': 'Maryland', 
          'AK': 'Alaska', 'UT': 'Utah', 'OR': 'Oregon', 'MT': 'Montana', 'IL': 'Illinois',
          'TN': 'Tennessee', 'DC': 'District of Columbia', 'VT': 'Vermont', 'ID': 'Idaho', 
          'AR': 'Arkansas', 'ME': 'Maine', 'WA': 'Washington', 'HI': 'Hawaii', 'WI': 'Wisconsin',
          'MI': 'Michigan', 'IN': 'Indiana', 'NJ': 'New Jersey', 'AZ': 'Arizona', 'GU': 'Guam', 
          'MS': 'Mississippi', 'PR': 'Puerto Rico', 'NC': 'North Carolina', 'TX': 'Texas', 'SD': 'South Dakota',
          'MP': 'Northern Mariana Islands', 'IA': 'Iowa', 'MO': 'Missouri', 'CT': 'Connecticut', 
          'WV': 'West Virginia', 'SC': 'South Carolina', 'LA': 'Louisiana', 'KS': 'Kansas', 
          'NY': 'New York', 'NE': 'Nebraska', 'OK': 'Oklahoma', 'FL': 'Florida', 'CA': 'California', 
          'CO': 'Colorado', 'PA': 'Pennsylvania', 'DE': 'Delaware', 'NM': 'New Mexico', 
          'RI': 'Rhode Island', 'MN': 'Minnesota', 'VI': 'Virgin Islands', 'NH': 'New Hampshire', 
           'MA': 'Massachusetts', 'GA': 'Georgia', 'ND': 'North Dakota', 'VA': 'Virginia'})

    housing = pd.read_csv('/Users/emilyvincett/Downloads/City_Zhvi_AllHomes.csv')
    #housing.columns[8:56]
    housing.drop(housing.columns[8:56], axis=1, inplace=True)
    #housing.columns[209:251]
    housing.drop(housing.columns[209:251],axis=1,inplace=True)
    housing['State'] = housing['State'].replace(states)
    housing.drop(['SizeRank', 'RegionType','StateName','Metro','CountyName'], axis=1, inplace=True)

    def convert_to_qtr(ym):
        year,month,day = ym.split('-')
        if month == '01' or month == '02' or month == '03': 
            result = year + 'q1'
        elif month == '04' or month == '05' or month == '06': 
            result = year + 'q2'
        elif month == '07' or month == '08' or month == '09':       
            result = year + 'q3'
        else:
            result = year + 'q4'
        return result

    h2 = housing.copy().set_index(['State', 'RegionName','RegionID']).stack(dropna=False)
    #h2.head(100)
    h2 = h2.reset_index().rename(columns ={'level_3':'year_month',0: 'gdp'})
    h2.drop_duplicates(inplace=True)
    h2['quarter'] = h2['year_month'].apply(convert_to_qtr)
    h2 = h2.drop('year_month',axis=1)
    result = h2.pivot_table(values='gdp', index=['State', 'RegionName', 'RegionID'], columns='quarter', aggfunc=np.mean)
    result = result.reset_index()
    result = result.drop('RegionID', axis=1)
    result = result.set_index(['State', 'RegionName'])
    return result

convert_housing_data_to_quarters()
```
