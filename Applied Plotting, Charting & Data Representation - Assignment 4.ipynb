```python
%matplotlib notebook
import pandas as pd
import numpy as np 
import matplotlib as mpl
import matplotlib.pyplot as plt


house_prices = pd.read_excel('/Users/emilyvincett/Downloads/ratioofhousepricetoworkplacebasedearnings.xlsx', 
                             '1a',skiprows=6)
house_prices = house_prices[0:14]
house_prices.drop('Code',axis=1,inplace=True)
house_prices.reset_index(drop=True)
house_prices = house_prices.set_index('Name')
house_prices = house_prices.dropna()
house_prices = house_prices.reset_index()
house_prices['% House Price Growth'] = (house_prices['Year ending Sep 2019']/house_prices['Year ending Sep 1997'])*100
house_prices.sort_values('% House Price Growth',ascending=False)
house_prices.reset_index(drop=True)
house_prices = house_prices.set_index('Name')
cols_to_rename = {'Year ending Sep 1997':'97', 'Year ending Sep 1998':'98', 'Year ending Sep 1999':'99',
                'Year ending Sep 2000':'00', 'Year ending Sep 2001':'01', 'Year ending Sep 2002':'02',
                'Year ending Sep 2003':'03', 'Year ending Sep 2004':'04', 'Year ending Sep 2005':'05',
               'Year ending Sep 2006':'06', 'Year ending Sep 2007':'07', 'Year ending Sep 2008':'08',
               'Year ending Sep 2009':'09', 'Year ending Sep 2010':'10', 'Year ending Sep 2011':'11',
               'Year ending Sep 2012':'12', 'Year ending Sep 2013':'13', 'Year ending Sep 2014':'14',
               'Year ending Sep 2015':'15', 'Year ending Sep 2016':'16', 'Year ending Sep 2017':'17',
               'Year ending Sep 2018':'18', 'Year ending Sep 2019':'19'}
house_prices.rename(cols_to_rename, axis=1,inplace=True)
house_prices.drop('% House Price Growth',axis=1,inplace=True)
house_prices
```

```python
median_income = pd.read_excel('/Users/emilyvincett/Downloads/ratioofhousepricetoworkplacebasedearnings.xlsx',
                              '1b',skiprows=6)
median_income.drop('Unnamed: 4',axis=1,inplace=True)
median_income = median_income[0:14]
median_income.drop('Code',axis=1,inplace=True)
median_income = median_income.dropna()
median_income.columns = median_income.columns.astype(str)
median_income['% Income Increase'] = (median_income['2019']/median_income['1997'])*100
median_income = median_income.set_index('Name')
median_income.drop('% Income Increase',axis=1,inplace=True)
cols_to_rename = ({'1997':'97','1998':'98','1999':'99',
                   '2000':'00','2001':'01','2002':'02',
                   '2003':'03','2004':'04','2005':'05',
                   '2006':'06','2007':'07','2008':'08',
                   '2009':'09','2010':'10','2011':'11',
                   '2012':'12','2013':'13','2014':'14',
                   '2015':'15','2016':'16','2017':'17',
                   '2018':'18','2019':'19'})
median_income.rename(cols_to_rename,axis=1,inplace=True)
median_income
```

```python
ratio = median_income/house_prices
ratio_e_w = ratio.loc['England and Wales']

e_w_house_prices = house_prices.loc['England and Wales']
#e_w_house_prices.index
e_w_income = median_income.loc['England and Wales']
fig = plt.figure()
ax1 = plt.subplot(1,2,1)
plt.plot(e_w_house_prices.index,e_w_house_prices,label='House Prices: 1997 - 2019',
         color='blue',linestyle='dotted')
plt.xlabel('Year')

y = plt.gca().yaxis
for item in y.get_ticklabels():
    item.set_fontsize(9)
    item.set_rotation(45)

x = plt.gca().xaxis
for item in x.get_ticklabels():
    item.set_rotation(90)
    item.set_fontsize(6.5)
plt.title('UK House Prices in GBP',size=8.5)


ax2 = plt.subplot (1,2,2)
plt.plot(ratio_e_w.index, ratio_e_w, label='Ratio = Incomes/House Prices',color='red',linestyle='dotted')
plt.legend(loc=3,fontsize=7)
plt.xlabel('Year')

y = plt.gca().yaxis
for item in y.get_ticklabels():
    item.set_fontsize(9)
    item.set_rotation(45)

x = plt.gca().xaxis
for item in x.get_ticklabels():
    item.set_fontsize(6.5)
    item.set_rotation(90)

plt.title('Incomes to House Price Ratio',size=8.5)
fig.savefig('housing_and_incomes.png',dpi=fig.dpi)
```
