```python
import pandas as pd
df = pd.read_csv('/Users/emilyvincett/Downloads/olympics.csv',index_col=0,skiprows=1)

for col in df.columns:
    if col[:2]=='01':
        df.rename(columns={col:'Gold'+col[4:]}, inplace=True)
    if col[:2]=='02':
        df.rename(columns={col:'Silver'+col[4:]}, inplace=True)
    if col[:2]=='03':
        df.rename(columns={col:'Bronze'+col[4:]}, inplace=True)
    if col[:1]=='№':
        df.rename(columns={col:'#'+col[1:]}, inplace=True)

names_ids = df.index.str.split('\s\(') # split the index by '('

df.index = names_ids.str[0] # the [0] element is the country name (new index) 
df['ID'] = names_ids.str[1].str[:3] # the [1] element is the abbreviation or ID (take first 3 characters from that)

df = df.drop('Totals')

df.head()

```

```python
def answer_one():
    return df['Gold'].idxmax()

answer_one()
```

```python
def answer_two():
    return (df['Gold'] - df['Gold.1']).abs().idxmax()

answer_two()
```

```python
def answer_four():
    GP = df['Gold.2']*3
    SP = df['Silver.2']*2
    BP = df['Bronze.2']* 1
    Points = GP + SP + BP
    return Points

answer_four()
```

```python
def answer_three():
    swg_df = df.copy()
    swg_df = swg_df[(swg_df['Gold']>0) & (swg_df['Gold.1']>0)]
    swg_df['diff'] = (swg_df['Gold'] - swg_df['Gold.1']).abs()
    swg_df['tot'] = swg_df['Gold'] + swg_df['Gold.1']
    swg_df['relative'] = swg_df['diff']/swg_df['tot']
    return swg_df['relative'].idxmax()

answer_three()
```

```python
c_df = pd.read_csv('/Users/emilyvincett/Downloads/census.csv')
c_df.head()
```

```python
def answer_five():
    c_df1 = c_df.copy()
    c_df1 = c_df1[c_df1['SUMLEV']==50]
    c_df1 = c_df1.reset_index()
    c_df1 = c_df1.set_index(['STNAME','CTYNAME'])
    c_df1 = c_df1.reset_index(level=1)
    return c_df1.index.value_counts().idxmax()

answer_five()
```

```python
def answer_six():
    c_df2 = c_df.copy()
    c_df2 = c_df2[c_df2['SUMLEV']==50]
    needed_columns = ['STNAME','CTYNAME','CENSUS2010POP']
    c_df2 = c_df2[needed_columns]
    #c_df2.head()
    c_df2 = c_df2.sort_values(['STNAME','CENSUS2010POP'],ascending=False)
    c_df2 = c_df2.groupby('STNAME')
    c_df2 = c_df2.head(3)
    c_df2 = c_df2.nlargest(3,'CENSUS2010POP')
    c_df2 = c_df2.reset_index()
    return list(c_df2.iloc[0:3]['STNAME'])

answer_six()
```

```python
def answer_seven():
    c_df3 = c_df.copy()
    c_df3 = c_df3[c_df3['SUMLEV']==50]
    needed_columns = ['CTYNAME','POPESTIMATE2010',
                  'POPESTIMATE2011','POPESTIMATE2012',
                  'POPESTIMATE2013','POPESTIMATE2014','POPESTIMATE2015']
    c_df3['maxest'] = c_df3[needed_columns].max(axis=1)
    c_df3['minest'] = c_df3[needed_columns].min(axis=1)
    c_df3['change'] = (c_df3['maxest']- c_df3['minest']).abs()
    kept_columns = ['CTYNAME','change']
    c_df3 = c_df3[kept_columns]
    c_df3 = c_df3.nlargest(1,'change')
    c_df3 = c_df3.reset_index()
    return c_df3.iloc[0]['CTYNAME']

answer_seven()
```

```python
def answer_eight():
    c_df4 = c_df.copy()
    c_df4 = c_df4[c_df4['SUMLEV']==50]
    kept_columns = ['STNAME','CTYNAME','REGION','POPESTIMATE2015','POPESTIMATE2014']
    c_df4 = c_df4[kept_columns]
    c_df4 = c_df4[(c_df4['REGION']<= 2)]
    c_df4 = c_df4[c_df4['POPESTIMATE2015']>c_df4['POPESTIMATE2014']]
    c_df4 = c_df4[c_df4['CTYNAME'].str.startswith('Washington')]
    kept_columns_2 = ['STNAME','CTYNAME']
    return c_df4[kept_columns_2]

answer_eight()
```
